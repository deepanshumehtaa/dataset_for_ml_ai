```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

url = 'https://raw.githubusercontent.com/deepanshumehtaa/csvfiles/master/haberman.csv'
df = pd.read_csv(url, index_col=0, delimiter=',')
df = df.reset_index()
df.head(15)
```

```python
sb.heatmap( df, cmap = 'magma' )
plt.show()

df.info()
```

```python
# Age of patient at time of operation


def uni(X,Y,da):                       # for univariant
    plt.figure(figsize = (10,5))
    sb.barplot(x= X, y =Y, data = da)
    plt.show()

def pp(da):
    plt.figure(figsize = (10,5))
    sb.pairplot(data = da)
    plt.show()
    
def bi(X,Y,da):                         # for bivariant analysis
    plt.figure(figsize = (10,5))        # for joint plot we need to give x and y in ""
                                        # same for  barplot()
    #sb.jointplot(x= X, y =Y, data = da)
    sb.jointplot(x='Age', y="Op_Year", data=da)
    plt.show()
    
def box(X,Y,da):                         # for bivariant analysis
    plt.figure(figsize = (10,5))
    sb.boxplot(x= X, y =Y, data = da)
    plt.show()
    
def vio(X,Y,da):                         # for bivariant analysis
    plt.figure(figsize = (10,5))
    sb.violinplot(x= X, y =Y, data = da)
    plt.show()
    
def meann(X,Y,da):                         # for bivariant analysis
    plt.figure(figsize = (10,5))
    sb.violinplot(x= X, y =Y, data = da)
    plt.show()
    
def stdd(X,Y,da):                         # for bivariant analysis
    plt.figure(figsize = (10,5))
    sb.violinplot(x= X, y =Y, data = da)
    plt.show()
    
def varr(X,Y,da):                         # for bivariant analysis
    plt.figure(figsize = (10,5))
    sb.violinplot(x= X, y =Y, data = da)
    plt.show()
    
def vio(X,Y,da):                         # for bivariant analysis
    plt.figure(figsize = (15,5))
    sb.violinplot(x= X, y =Y, data = da)
    plt.show()
    

x= df.Age 
y = df.Op_Year
data = df

uni(x,y,data)
pp(data)
bi(x,y,data)  #bi-variant plot
box(x,y,data)
vio(x,y,data)
```
